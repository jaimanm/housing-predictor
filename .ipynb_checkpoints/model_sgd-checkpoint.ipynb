```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDRegressor
```

```python
# load the data
df = pd.read_csv('california_housing.csv')
df.head()
```

```python
X = df.drop(labels=['median_house_value', 'ocean_proximity'], axis=1)
y = pd.DataFrame(df['median_house_value'])
X, y
```

```python
# plot data
_, n = X.shape
fig, ax = plt.subplots(1, n, sharey=True, figsize=(25, 3))
for i in range(n):
    ax[i].scatter(X[X.columns[i]], y)
    ax[i].set_xlabel(X.columns[i])   
    
ax[0].set_ylabel(y.columns[0])

plt.show()
```
